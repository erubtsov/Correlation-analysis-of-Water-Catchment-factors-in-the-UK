This data set is continuous measurements of conductivity, dissolved oxygen, pH, temperature and water level from the Frome Piddle; Pang Lambourn and Tern catchments, recorded between 2002 and 2007. These are rivers in the United Kingdom and this data was supplied by the government of the United Kingdom.

Reading the data and parsing the dates:

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Latitude E7470\Documents\DSCI 235 FInal Project Data Set\00895096-e083-498f-9a52-b7f9e3291125\data\LCWQ_Continuous_PLSites.csv",index_col='Date', 
                       parse_dates=True)
df
```

This is a huge data set, so I want to contain it to yearly measurements. I break up the time index into a day, month, and year column. I do this for all 3 locations. I take the average of every column to attribute to each year.

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df_grouped = df.groupby('year')['Battery (V)', 'Conductivity 25C continuous (uS/cm)', 'Oxygen dissolved continuous (%satn)', 'pH continuous', 'Temperature water continuous (C)', 'Water level continuous (mm)'].mean()
df_grouped
```

```python
df1['day'] = df1.index.day
df1['month'] = df1.index.month
df1['year'] = df1.index.year
df1_grouped = df1.groupby('year')['Battery (V)', 'Conductivity 25C continuous (uS/cm)', 'Oxygen dissolved continuous (%satn)', 'pH continuous', 'Temperature water continuous (C)', 'Water level continuous (mm)'].mean()
df1_grouped
```

```python
df2 = pd.read_csv(r"C:\Users\Latitude E7470\Documents\DSCI 235 FInal Project Data Set\00895096-e083-498f-9a52-b7f9e3291125\data\LCWQ_Continuous_TESites.csv",index_col='Date', 
                       parse_dates=True)
df2
```

```python
df2['day'] = df2.index.day
df2['month'] = df2.index.month
df2['year'] = df2.index.year
df2_grouped = df2.groupby('year')['Battery (V)', 'Conductivity 25C continuous (uS/cm)', 'Oxygen dissolved continuous (%satn)', 'pH continuous', 'Temperature water continuous (C)', 'Water level continuous (mm)'].mean()
df2_grouped
```

Here, I concatenate the data frames and specify their locations.

```python
water_attributes = ['Battery (V)', 'Conductivity 25C continuous (uS/cm)', 'pH continuous', 
                     'Temperature water continuous (C)', 'Temperature water continuous (C)', 'Water level continuous (mm)']
dfg = round(df_grouped[water_attributes].describe(),3)
df1g = round(df_grouped[water_attributes].describe(),3)
df2g = round(df2_grouped[water_attributes].describe(),3)

c = pd.concat([dfg, df1g, df2g], axis=1, keys=['Location 1', 'Location 2', 'Location 3'])
c
```

Finally, let's look at the correlation that the different values in locations have among eachother using a heatmap from the seaborn package:

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(10, 6))
corr = c.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Water Attributes Correlation Heatmap', fontsize=14)
```
